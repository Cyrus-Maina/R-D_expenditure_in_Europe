This dataset represents Goverment expenditure by EU coutries on R&D, sourced from Kaggle.

R&D is a form of fiscal policy tool for the Goverment to expand or contract the Economy.

R&D births innovation which leads to improved quality and quantity of Factors of Production which blesses an economy with sustainable Long Run Aggregate Supply

E.g., R&D in the Higher Education looks like the Goverment giving grants to a University to develop/improve robotic tehnology or in the Private Enterprise, giving grants to Tesla to improve car battery performance.

This in turn makes US dominant in the EV business as Tesla will be way ahead of other nations not subsidizing their EV industry and have to import from US, these high performance batteries.

This dataframe has 5 fields and 2556 rows.
The Geo field represents a Countries name abbreviated by abbreviations e.g., AT for Austria

Our years start from 2008-2019.

'sectperf' represents economy sectors performance.
BES represents Business Sector
GOV represents Goverment sector
HES represents Higher Education Sector

```python
# Filter data for Austria ('AT')
austria_data = df[(df['geo'] == 'AT') & (df['unit'] == 'PC_GDP')]
# Group the data by 'sectperf' and 'year' and calculate the mean value for each sector
grouped_data = austria_data.groupby(['sectperf', 'year']).mean().reset_index()
# Pivot the data for easy plotting
pivot_data = grouped_data.pivot(index='year', columns='sectperf', values='value')
# Plotting
plt.figure(figsize=(10, 6))
pivot_data.plot(kind='line', marker='o', linestyle='-')
plt.title('R&D Expenditure by Sector in Austria (2008-2019)')
plt.xlabel('Year')
plt.ylabel('R&D Expenditure (% of GDP)')
plt.legend(title='Sector', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

The Private enterprise has been receiving increasing grants across the years consistently.
From this graph we can tell Austria's goal is to expand the economy.
Its allocated, private eterprise, more than 2% of total GDP.
Austria's Goverment seems to prioritize improved ways of production which it hopes will spur gradual economic growth.

Think grants to Redbull to innovate on their factors of production and come up with new beverages which will earn the nation more revenues and reduced unemployment leves

Think R&D grants to GLOCK GmbH, handgun manufacturer to innovate on more efficient guns which will lead to more sales and increased revenues from exports, strong currecy from the exports, upskilling of labor force, decreased unemployment which leads to increased cosumption.

The goverment then spends almost a constant 1% of its GDP on Higher Education to spur upskilling of its upcoming labor force, create new ways of manufacturing that are cost efficient e.g., improved/new ways of mechanical engineering that may be applied in firms like Glock GMBH.

All these assumptions point to how Goverment spends via R&D to expand or contract the economy.

```python
#let's analyze a powerhouse like Germany 'DE'
germany_data = df[(df['geo'] == 'DE') & (df['unit'] == 'PC_GDP')]

grouped_data = germany_data.groupby(['sectperf', 'year']).mean().reset_index()
# Pivot the data for easy plotting
pivot_data = grouped_data.pivot(index='year', columns='sectperf', values='value')
# Plotting
plt.figure(figsize=(10, 6))
pivot_data.plot(kind='line', marker='o', linestyle='-')
plt.title('R&D Expenditure by Sector in Germany (2008-2019)')
plt.xlabel('Year')
plt.ylabel('R&D Expenditure (% of GDP)')
plt.legend(title='Sector', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Unlike Austria, Germany spends nothing on Non-profit sectors.

It however channels more than 2% of GDP in recent years to the Private Enterprise Sector. The spending has also been increasing signifying Economy Expansion efforts.

Germany's GDP in 2019 was 3.8 Trillion USD, 2% of that means 77 Billion USD channeled towards R&D!

Say, R&D grants to Volkswagen will help them innovate on their KPIs say higher performace cars and electric vehicles  which may increase sales in foreseeable future.

This buzz of innovation will lead to a need of new skills, which translates to an upskilling of the labor force and university students which means improved quality and quantity of Human capital side of production in the long term.

These new skills will fetch good incomes which will be disposed into the economy. Increased EVs production will lead to high exports strengthening the Euro.
Therefore R&D is a key fiscal tool for manipulating the economy.

```python
#LET'S plot a % change in R&D expenditure across the years
import matplotlib.pyplot as plt
germany_data = df[(df['geo'] == 'DE') & (df['unit'] == 'PC_GDP')]
# Calculate percentage change from the previous year
germany_data['percentage_change'] = germany_data['value'].pct_change() * 100
# Plotting percentage change in R&D expenditure as a percentage of GDP for Germany
plt.figure(figsize=(8, 6))
plt.plot(germany_data['year'], germany_data['percentage_change'], marker='o', linestyle='-')
plt.title('Percentage Change in R&D Expenditure (% of GDP) in Germany (Yearly)')
plt.xlabel('Year')
plt.ylabel('Percentage Change')
plt.grid(True)
plt.tight_layout()
plt.show()
```

From the plot, we can tell  Angela Merkel has slightly been expanding and contracting R&D expenditure.

The % changes are very minute thus we can say the Chancellor  prioritized improving the quality and quantity of factors of production.

We can say it worked as the GDP has grown from 2.8 Trillion USD when she entered office to 4.26 trillion USD a 64.29% increase.

The economy grew by 64% in her reign thus we can say her Fiscal policy on Goverment spending helped meet the macroeconomic goal of economy growth.

We can average annual economy growth to 4% that is 64% divided by her 16 years of her reign(assuming)

```python
#let's plot R&D expenditure in EU to the various sectors across the years.
eu_countries = ['AT', 'BE', 'BG', 'CY', 'CZ', 'DE', 'DK', 'EE', 'ES', 'FI', 'FR', 'GR', 'HR', 'HU', 'IE', 'IT', 'LT', 'LU', 'LV', 'MT', 'NL', 'PL', 'PT', 'RO', 'SE', 'SI', 'SK']
eu_data = df[(df['geo'].isin(eu_countries)) & (df['unit'] == 'PC_GDP')]
grouped_eu_data = eu_data.groupby(['sectperf', 'year']).mean().reset_index()
pivot_eu_data = grouped_eu_data.pivot(index='year', columns='sectperf', values='value')
plt.figure(figsize=(10, 6))
pivot_eu_data.plot(kind='line', marker='o', linestyle='-')
plt.title('Average R&D Expenditure by Sector in the EU (2008-2019)')
plt.xlabel('Year')
plt.ylabel('Average R&D Expenditure (% of GDP)')
plt.legend(title='Sector', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Total R&D expenditure has been increasing with the exception of 2014-2015.
This reflects the EU's desire to expand their economies.
Most of the R&D expenditure is flowing into the Business sector with an average 1% of the GDP.
This is with reasons:
1. Business sector has a quicker return on investment like  increase in supply where as educatio requires tons of time to realize impact. Educataion is more long term.

2. Expanded businesses lower unemployment, increase revenues and expands economy.

3. The funds will help industries work at full capacity. This means businesses will optimize all resources.

Non profits received the lowest cut as their contributions are not perceived to have a direct impact on the economy i.e., expanding or contracting.
If anything their allocation has been dwindling across the years.

We can also compare different Goverment spendings. Though we should be careful of the flawed state of such an analysis.

1% of Germany's 3.88 Trillion USD GDP is much more than 1.5% of Estonia's 31 Billion dollar economy in 2019.

The commitment is higher but cash size different.

```python
rd_expenditure = df[df['unit'] == 'PC_GDP']
average_expenditure = rd_expenditure.groupby('geo')['value'].mean().sort_values(ascending=False)
print(average_expenditure.head(10))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
average_expenditure.head(10).sort_values().plot(kind='bar', color='skyblue')
plt.title('Average R&D Expenditure (PC_GDP) by Country')
plt.xlabel('Country')
plt.ylabel('Average R&D Expenditure (% of GDP)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.ylim(0,1.8)
plt.show()
```

Germany, Sweden, Denmark are among the countries that allocate the largest size of GDP to R&D expenditure.
These countries also have large GDPs e.g., Denmark with a GDP of 393 Billion USD in 2023.

Let's also analyze which countries allocate the highest average GDP%  to R&D.
This will reflect countries prioritizing Human Capital as an improved factor of production.

```python
hes_expenditure = df[(df['unit'] == 'PC_GDP') & (df['sectperf'] == 'HES')]
highest_hes_expenditure = hes_expenditure.groupby('geo')['value'].mean().sort_values(ascending=False)
highest_hes_expenditure.head()
```
